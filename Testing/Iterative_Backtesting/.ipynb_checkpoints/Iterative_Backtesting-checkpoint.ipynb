A First Intuition on Iterative Backtesting (Part2)

```python
data
```

```python
sma_s = 50
sma_l = 200
```

```python
data["SMA_S"] = data.price.rolling(sma_s).mean()
data["SMA_L"] = data.price.rolling(sma_l).mean()
```

```python
data.dropna(inplace = True)
```

```python
data
```

```python
position = 0
```

```python
len(data)
```

```python
for bar in range(len(data)):
    if data["SMA_S"].iloc[bar] > data["SMA_L"].iloc[bar]:
        if position in [0, -1]:
            print("{}: Go Long | Price: {} | Spread: {}".format(data.index[bar].date(), data.price[bar], data["SMA_S"].iloc[bar] - data["SMA_L"].iloc[bar]))
            position = 1
    elif data["SMA_S"].iloc[bar] < data["SMA_L"].iloc[bar]:
        if position in [0, 1]:
            print("{}: Go Short | Price: {} | Spread: {}".format(data.index[bar].date(), data.price[bar], data["SMA_S"].iloc[bar] - data["SMA_L"].iloc[bar]))
            position = -1
```

Creating an Iterative Base Class (Part1)

```python
class IterativeBase(): #base class not dependent on the strategy type. The following is need for all of them. 
    def __init__(self, symbol, start, end, amount):
        self.symbol = symbol
        self.start = start
        self.end = end
        self.initial_balance = amount
        self.current_balance = amount
        self.get_data()

    def get_data(self):
        raw = pd.read_csv("detailed.csv", parse_dates = ["time"], index_col = "time").dropna()
        raw = raw.loc[self.start:self.end]
        raw["returns"] = np.log(raw.price / raw.price.shift(1))
        self.data = raw

    def plot_data(self, cols = None):
        if cols is None:
            cols = "price"
        self.data[cols].plot(figsize = (12, 8), title = self.symbol)
```

```python
class IterativeBase(): #base class not dependent on the strategy type. The following is need for all of them. 
    def __init__(self, symbol, start, end, amount):
        self.symbol = symbol
        self.start = start
        self.end = end
        self.initial_balance = amount
        self.current_balance = amount
        self.get_data()

    def get_data(self):
        raw = pd.read_csv("detailed.csv", parse_dates = ["time"], index_col = "time").dropna()
        raw = raw.loc[self.start:self.end]
        raw["returns"] = np.log(raw.price / raw.price.shift(1))
        self.data = raw

    def plot_data(self, cols = None):
        if cols is None:
            cols = "price"
        self.data[cols].plot(figsize = (12, 8), title = self.symbol)

    def get_values(self, bar): # this is the new methode added
        data = str(self.data.index[bar].date())
        price = round(self.data.price.iloc[bar], 5)
        spread = round(self.data.spread.iloc[bar], 5)
        return data, price, spread
```

Iterative Base Class add Buy Function

```python
class IterativeBase(): #base class not dependent on the strategy type. The following is need for all of them. 
    def __init__(self, symbol, start, end, amount):
        self.symbol = symbol
        self.start = start
        self.end = end
        self.initial_balance = amount
        self.current_balance = amount
        self.units = 0
        self.trades = 0
        self.get_data()

    def get_data(self):
        raw = pd.read_csv("detailed.csv", parse_dates = ["time"], index_col = "time").dropna()
        raw = raw.loc[self.start:self.end]
        raw["returns"] = np.log(raw.price / raw.price.shift(1))
        self.data = raw

    def plot_data(self, cols = None):
        if cols is None:
            cols = "price"
        self.data[cols].plot(figsize = (12, 8), title = self.symbol)

    def get_values(self, bar): # this is the new methode added
        data = str(self.data.index[bar].date())
        price = round(self.data.price.iloc[bar], 5)
        spread = round(self.data.spread.iloc[bar], 5)
        return data, price, spread

    def print_current_balance(self, bar):
        date, price, spread = self.get_values(bar)
        print("{} | Current Balance: {}".format(date, round(self.current_balance, 2)))

    def buy_instrument(self, bar, units = None, amount = None):
        date, price, spread = self.get_values(bar)
        if amount is not None:
            units = int(amount / price)
        self.current_balance -= units * price
        self.units += units
        self.trades += 1
        print("{} | Buying {} for {}".format(date, units, round(price, 5)))
```

```python
bc = IterativeBase("EURUSD", "2006-12-31", "2020-06-30", 100000)
```

```python
bc.print_current_balance(0)
```

```python
bc.data
```

```python
bc.buy_instrument(0, units = 1000)
```

```python
bc.units
```

```python
bc.print_current_balance(0)
```

```python
bc.buy_instrument(1 , amount = 5000)
```

```python
bc.print_current_balance(1)
```
