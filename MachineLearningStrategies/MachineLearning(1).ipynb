```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
data = pd.read_csv("Five_minute.csv", parse_dates = ["time"], index_col = "time")
```

```python
data
```

```python
data["returns"] = np.log(data.div(data.shift(1)))
```

```python
data.dropna(inplace = True)
```

```python
data["direction"] = np.sign(data.returns)
```

```python
data
```

```python
lags = 2
```

```python
cols = []
for lag in range(1, lags + 1):
    col = "lag{}".format(lag)
    data[col] = data.returns.shift(lag)
    cols.append(col)
data.dropna(inplace = True)
```

```python
data
```

```python
lm = LogisticRegression(C = 1e6, max_iter = 80000, multi_class = "ovr")
```

```python
lm.fit(data[cols], data.direction)
```

```python
lm
```

```python
data["pred"] = lm.predict(data[cols]) #predicting data
```

```python
data
```

```python
hits = np.sign(data.direction * data.pred).value_counts() # predicting value
```

```python
hits
```

```python
hit_ratio = hits[1.0] / sum(hits)
hit_ratio
```
