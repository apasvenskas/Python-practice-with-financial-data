# Error Handling

```python
my_list = [1, 2, 3, "Four", 5, 6, 7]
```

```python
for element in my_list:
    result = 10 + element
    print(result)
```

### Using try and except for debugging 

```python
for element in my_list:
    try:
        results = 10 + element
        print(result)
    except:
        print("{} is not a number.".format(element))
```

### Specific Error Handling

```python
for element in my_list:
    try:
        results = 10 + element
        print(result)
    except TypeError: 
        print("{} is not a number.".format(element))
```

```python
for element in my_list:
    try:
        results = 10 + element
        print(result)
    except ValueError:
        print("{} is not a number.".format(element))
```

```python
for element in my_list:
    try:
        results = 10 + element
        print(result)
    except Exception as e:
        print("{} is not a number.".format(element), end = " | ")
        print(e)
```
