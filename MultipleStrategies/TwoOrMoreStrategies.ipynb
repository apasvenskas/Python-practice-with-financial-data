Combininng Both Strategies

```python
tester.results
```

```python
tester2.results
```

```python
comb = tester.results.loc[:, ["returns", "position"]].copy()
```

```python
comb
```

```python
comb.rename(columns = {"position":"position_SMA"}, inplace = True)
```

```python
comb["position_MR"] = tester2.results.position.astype("int")
```

```python
comb
```

Methode 1

```python
comb["position_comb"] = np.where(comb.position_MR == comb.position_SMA, comb.position_MR, 0) # methode #1
```

```python
comb.head(60) # first 60 rows
```

```python
comb.position_comb.value_counts()
```

```python
comb.position_comb.plot(figsize = (12, 8))
plt.show()
```
